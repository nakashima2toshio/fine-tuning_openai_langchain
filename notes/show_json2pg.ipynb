```python
import json

# Load the JSON file
file_path = '../data/code/chatbot.json'

with open(file_path, 'r') as file:
    data = json.load(file)

# Print the program names and their codes
for program_name, program_code in data.items():
    print(f"# Program Name: {program_name}")
    print("# Program Code:")
    print(program_code)
    print("\n" + "="*50 + "\n")
    
```
